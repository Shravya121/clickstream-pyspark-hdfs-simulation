```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Clickstream_Analysis_Simulated_HDFS") \
    .getOrCreate()
```

```python
file_path = "/content/drive/MyDrive/hdfs_sim/input/ecommerce_clickstream_transactions.csv"

df = spark.read.option("header", True).csv(file_path, inferSchema=True)
df.show(5)
df.printSchema()
```

```python
top_events = df.groupBy("EventType").agg(count("*").alias("event_count")).orderBy(desc("event_count"))
top_events.show(5)

```

```python
# Handle null or missing values
df = df.na.drop()

# Convert Timestamp to proper datetime
from pyspark.sql.functions import to_timestamp
df = df.withColumn("Timestamp", to_timestamp("Timestamp"))

# Extract date and hour (for time-based analysis)
from pyspark.sql.functions import hour, date_format
df = df.withColumn("hour", hour("Timestamp"))
df = df.withColumn("date", date_format("Timestamp", "yyyy-MM-dd"))
```

```python
from pyspark.sql.functions import hour, count

# Extract hour from timestamp and count events per hour
hourly_trend = (
    df.withColumn("hour", hour("timestamp"))
      .groupBy("hour")
      .agg(count("*").alias("events_per_hour"))
      .orderBy("hour")
)

# Now convert to pandas and plot
import matplotlib.pyplot as plt

pdf = hourly_trend.toPandas()
plt.plot(pdf['hour'], pdf['events_per_hour'], marker='o')
plt.title("User Activity per Hour")
plt.xlabel("Hour of Day")
plt.ylabel("Number of Events")
plt.grid(True)
plt.show()

```

```python
df.write.mode("overwrite").parquet("/content/drive/MyDrive/hdfs_sim/output/cleaned_clickstream")
```
